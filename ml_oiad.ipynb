```python
!pip install pymorphy2
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pymorphy2
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,multilabel_confusion_matrix,classification_report,f1_score
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_excel('/content/comments_ml.xlsx')[['text','Marker']]
df
```

```python
y = df['Marker'].apply(lambda x: x.upper())
X = df['text']
X
```

```python
import re
def preproc_text(text):
    norm_text=[]
    morph=pymorphy2.MorphAnalyzer()
    for s in text:
      try:
        s1 = re.sub(r'[^\w\s]+|[\d]+', r'',s).strip()
        s1 = s1.lower()
        s1 = word_tokenize(s1)
        words=[]
        for i in s1:
            pv = morph.parse(i)
            words.append(str(pv[0].normal_form))
        sentence=' '.join(words)
        norm_text.append(sentence)
      except Exception as e:
        norm_text.append('')
    return norm_text
```

```python
text_token = preproc_text(X)

```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=500,stop_words=stopwords.words('russian'))
X_desc = tfidf_vectorizer.fit_transform(text_token)
text_cv = pd.DataFrame(X_desc.toarray(),columns=tfidf_vectorizer.get_feature_names_out())
text_cv
```

```python
label = LabelEncoder()
y_label = label.fit_transform(y)
```

```python
text_cv['Y']=y_label
text_cv
```

```python
# 2 - P
# 1 - NT
# 0 - N
```

```python
X_train, X_test, y_train, y_test = train_test_split(text_cv.drop(columns=['Y']), text_cv['Y'], test_size=0.2, random_state=42)
```

```python
#model = RandomForestClassifier(n_estimators = 1000,max_depth=100)

model = RandomForestClassifier(n_estimators = 5000,max_depth=100)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
multilabel_confusion_matrix(y_test,y_pred)
```

```python
import warnings
#для удаления комментариев об ошибке в терминале
warnings.filterwarnings('ignore')
print(classification_report(y_test,y_pred))
```

```python
!pip install natasha
from natasha import (
    Segmenter,
    MorphVocab,
    PER,
    NamesExtractor,
    NewsNERTagger,
    NewsEmbedding,
    Doc
)
```

```python

emb = NewsEmbedding()
segmenter = Segmenter()
morph_vocab = MorphVocab()
ner_tagger = NewsNERTagger(emb)
names_extractor = NamesExtractor(morph_vocab)
new_X = []
for i in X:
  try:
    text_izm = i
    doc = Doc(i)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)
    for span in doc.spans:
        span.normalize(morph_vocab)
    {_.text: _.normal for _ in doc.spans}

    for span in doc.spans:
        if span.type == PER:
            span.extract_fact(names_extractor)
    for _ in doc.spans:
      if _.type == PER:
        text_izm = text_izm.replace(_.normal,'')
    new_X.append(text_izm)
  except TypeError:
      new_X.append('')
```

```python
len(new_X)
```

```python
text_token_new = preproc_text(new_X)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000,stop_words=stopwords.words('russian'),max_df=0.9)
X_desc = tfidf_vectorizer.fit_transform(text_token_new)
text_cv_new = pd.DataFrame(X_desc.toarray(),columns=tfidf_vectorizer.get_feature_names_out())
text_cv_new
```
